```python
import pandas as pd
import numpy as np
import re
```

```python
bond = pd.read_csv('tokenized_bond_results.csv')
mpb = pd.read_csv('tokenized_MPB_results.csv')
```

```python
bond.head()
```

```python
mpb.head()
```

```python
mpb = mpb.drop(['Unnamed: 0'], axis = 1)
mpb.head()
```

```python
word_list = []
total_words = []
cnt = 1
for string in bond['tagged']:
    words = re.findall(r"\('(.+?)',\s*'.+?'\)", string)
    word_list.append(words)
    total_words += words
    cnt += 1
```

```python
df_total_bond = pd.DataFrame(total_words, columns=['Word'])
df_total_bond
```

```python
counts = df_total_bond.value_counts()
counts.to_csv('bond_words.csv')
```

```python
word_list = []
total_words = []
cnt = 1
for string in mpb['tagged']:
    words = re.findall(r"\('(.+?)',\s*'.+?'\)", string)
    word_list.append(words)
    total_words += words
    cnt += 1
```

```python
df_total_mpb = pd.DataFrame(total_words, columns=['Word'])
df_total_mpb
```

```python
counts = df_total_mpb.value_counts()[0:100]
counts.to_csv('mpb_words.csv')
```
