```python
import requests
from bs4 import BeautifulSoup
import csv
import time

# 결과를 저장할 CSV 파일 이름
csv_filename = "news_data_test.csv"

# CSV 파일에 헤더 추가
with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Content", "Date"])

# 기본 URL
base_url = "https://biz.heraldcorp.com/search/detail?page={}&q=%EA%B8%88%EB%A6%AC&stype=NEWS&sort=&pageTop=true#pagetop"

# 페이지를 순차적으로 크롤링 (1페이지부터 5페이지까지)
for page_num in range(1, 6):  # 1부터 5까지 페이지 크롤링
    url = base_url.format(page_num)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # 각 기사의 제목, 내용, 날짜를 담고 있는 부분을 정확하게 찾기
    titles = soup.find_all('p', class_='news_title')  # 제목
    contents = soup.find_all('p', class_='news_text ellipsis2')  # 내용
    dates = soup.find_all('span', class_='date')  # 날짜

    # 각 기사에서 제목, 내용, 날짜를 추출하여 CSV 파일에 저장
    for title, content, date in zip(titles, contents, dates):
        title_text = title.get_text(strip=True)  # 제목
        content_text = content.get_text(strip=True)  # 내용
        date_text = date.get_text(strip=True) if date else 'No date'  # 날짜 (없으면 'No date')

        # 제목, 내용, 날짜 출력 (디버깅용)
        print(f"Title: {title_text}, Content: {content_text}, Date: {date_text}")

        # CSV 파일에 저장
        with open(csv_filename, mode='a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([title_text, content_text, date_text])

    print(f"Page {page_num} completed.")  # 진행 상황 출력

    # 서버 과부하 방지를 위해 잠시 대기
    time.sleep(1)  # 1초 대기

print("Crawling finished. Data saved to", csv_filename)
```
