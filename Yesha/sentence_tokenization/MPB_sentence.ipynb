```python
mpb_df = pd.read_csv(r"c:/Users/iq750/bootcamp_git/대용량 파일 관리/MPB.csv")
```

```python
# 문장 단위 토큰화 1. Kiwi
articles = mpb_df['article']
sentences = []
for article in articles:
    sent = [s.text for s in kiwi.split_into_sents(article)]
    sentences.append(sent)
mpb_df['sentences'] = sentences
```

```python
mpb_df.to_csv(r"c:/Users/iq750/bootcamp_git/대용량 파일 관리/MPB_sentences.csv")
```
